**Exercise 4.2** The numbers generated by random.random are supposed to be uniform between 0 and 1; that is, every value in the range should have the same probability.

Generate 1000 numbers from random.random and plot their PMF and CDF. Is the distribution uniform?

**IMPORTANT!** [ThinkStats2](https://github.com/AllenDowney/ThinkStats2) should be cloned ../.. relative to this dir.

```python
import sys
import os
import math

import numpy as np
import pandas as pd
```

```python
import thinkstats2_path
import thinkstats2, thinkplot
```

```python
rand1000 = np.random.random(1000)
```

```python
pmf = thinkstats2.Pmf(rand1000, label='pmf')
thinkplot.Pmf(pmf)
thinkplot.Show(xlabel='Random Numbers 0..1', ylabel='PMF')
```

```python
cdf = thinkstats2.Cdf(pmf, label='cdf')
thinkplot.Cdf(cdf)
thinkplot.Show(xlabel='Random Numbers 0..1', ylabel='CDF')
```

**Conclusion:** PMF plots are not suited for large number of values because the noise obscures the fact most values have the same probability. However the CDF plot clearly shows the uniform distrbution.
